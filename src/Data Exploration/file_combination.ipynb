```python
dataframes = {}

for file in os.listdir("reach example"):
    if file.endswith(".csv"):  # Ensure it's a CSV file
        file_path = os.path.join("reach example", file)
        df_name = f"{file[:-4]}_df"
        dataframes[df_name] = pd.read_csv(file_path)
```

```python
# info on data in each df
for df in dataframes:
    print(df)
    print(dataframes[df].shape)
```

```python
# Assuming 'dataframes' is your dictionary of DataFrames
single_row_dfs = [
    "trial_num_reach1_df", "reach_num_reach1_df", "cue_on_reach1_df", "reach_st_reach1_df", "reach_end_reach1_df",  
    "reach_pos_st_reach1_df", "reach_pos_end_reach1_df", "reach_dir_reach1_df", "reach_len_reach1_df"   
]

# Concatenate single-row DataFrames horizontally
trial_info_df = pd.concat([dataframes[df] for df in single_row_dfs], axis=1)
```

```python
trial_info_df
```

```python
# Assuming 'dataframes' is your dictionary of DataFrames
# only one left is "time_window_reach1_df", which just has which time windows correspond to the reach, this should already be in time stamps though
time_dfs = [
    "timestamps_reach1_df", "neural_data_PMd_reach1_df", 
     "reach1_df", "target_on_reach1_df"
]

# Concatenate single-row DataFrames horizontally
time_bin_df = pd.concat([dataframes[df] for df in time_dfs], axis=1)
```

```python
time_bin_df.head()
```

```python
# time stamps are in this as well
dataframes["reach1_df"].head()
```

```python
# verified that time_stamps and TimeStamp are identical
mismatch_indices = np.where(time_bin_df['TimeStamp'] != time_bin_df['time_stamps'])[0]
print(mismatch_indices)  # Prints indices of rows with non-identical values
```

```python
time_bin_df = time_bin_df.drop("time_stamps", axis=1)
```
