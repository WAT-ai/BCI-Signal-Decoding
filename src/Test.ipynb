```python
import nengo
from nengo.utils.ensemble import tuning_curves
import matplotlib.pyplot as plt
```

```python
model = nengo.Network()
with model:
    ens_test = nengo.Ensemble(50, dimensions=1)
with nengo.Simulator(model) as sim:
    pts, activity = tuning_curves(ens_test, sim)
plt.plot(pts, activity);
```
