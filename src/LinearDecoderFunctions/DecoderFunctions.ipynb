```python
import numpy as np
import pandas as pd
```

```python
data_df = pd.read_csv("merged_data.csv")
data_df
```

```python
n_neurons = 94
n_kinem = 4
```

```python
activity_matrix = data_df.loc[:, 'Neuron1':'Neuron94']
activity_matrix # n_timepoints x n_neurons
```

```python
tranposed_activity_matrix = activity_matrix.transpose()
tranposed_activity_matrix # n_neurons x n_timepoints
```

```python
kinem_matrix = data_df.loc[:, 'x_velocity':'y_acceleration']
kinem_matrix # n_timepoints x n_kinematics
```

```python
tranposed_kinem_matrix = kinem_matrix.transpose()
tranposed_kinem_matrix # n_kinematics x n_timepoints
```

```python
A = tranposed_activity_matrix.to_numpy() # n_neurons x n_timepoints
Y = tranposed_kinem_matrix.to_numpy() # n_kinematics x n_timepoints

D = np.linalg.lstsq(A @ A.T, A @ Y.T, rcond=None)[0].T
D_reg = np.linalg.lstsq(A @ A.T + A.shape[1] * np.square(np.random.normal(0,20,1)) * np.eye(n_neurons), A @ Y.T, rcond=None)[0].T
Y_hat = D @ A
Y_hat_reg = D_reg @ A
```

```python
def GetErrors(Y, Y_HAT):
    n = len(Y[0])

    # X VELOCITY RMSE
    x_velocity_rmse = np.sqrt(np.square(np.subtract(Y[0],Y_HAT[0])).mean())

    # Y VELOCITY RMSE
    y_velocity_rmse = np.sqrt(np.square(np.subtract(Y[1],Y_HAT[1])).mean())

    # X ACCELERATION RMSE
    x_acceleration_rmse = np.sqrt(np.square(np.subtract(Y[2],Y_HAT[2])).mean())

    # Y ACCELERATION RMSE
    y_acceleration_rmse = np.sqrt(np.square(np.subtract(Y[3],Y_HAT[3])).mean())

    return (x_velocity_rmse, y_velocity_rmse, x_acceleration_rmse, y_acceleration_rmse)
```

## Verifying Functional Decoder

```python
functional_error = GetErrors(Y, Y_hat)
print(f"X Velocity RMSE: {functional_error[0]}")
print(f"Y Velocity RMSE: {functional_error[1]}")
print(f"X Acceleration RMSE: {functional_error[2]}")
print(f"Y Acceleration RMSE: {functional_error[3]}")
```

## Verifying L2 Regression

```python
l2_error = GetErrors(Y, Y_hat_reg)
print(f"X Velocity RMSE: {l2_error[0]}")
print(f"Y Velocity RMSE: {l2_error[1]}")
print(f"X Acceleration RMSE: {l2_error[2]}")
print(f"Y Acceleration RMSE: {l2_error[3]}")
```
